```python
from dearpygui import dearpygui as dpg
import pandas as pd
import numpy as np
import uuid


dpg.create_context()

df_actual = None
color_actual = [0, 255, 0, 255]

# --- Funciones ---
def cargar_datos(sender, app_data):
    """Lee archivo CSV y actualiza combos y grfico"""
    global df_actual
    filepath = app_data['file_path_name']
    try:
        df_actual = pd.read_csv(filepath)
        columnas = df_actual.columns.tolist()
        dpg.configure_item("combo_x", items=columnas)
        dpg.configure_item("combo_y", items=columnas)
        actualizar_grafico(columnas[0], columnas[1])
    except Exception as e:
        print("Error al cargar datos:", e)

def actualizar_grafico(col_x, col_y):
    if df_actual is None:
        return
    try:
        x = df_actual[col_x].to_list()
        y = df_actual[col_y].to_list()
        dpg.set_value("serie_datos", [x, y])
        dpg.configure_item("serie_datos", show=True)
        dpg.set_item_label("serie_datos", f"{col_y} vs {col_x}")
    except Exception as e:
        print("Error al graficar:", e)

def cambiar_color(sender, app_data):
    dpg.configure_item("serie_datos", color=app_data)

def cambiar_grosor(sender, app_data):
    dpg.configure_item("serie_datos", weight=app_data)

def cambio_columna(sender, app_data):
    col_x = dpg.get_value("combo_x")
    col_y = dpg.get_value("combo_y")
    if col_x and col_y:
        actualizar_grafico(col_x, col_y)

def toggle_grid(sender, app_data):
    dpg.configure_item("x_axis", show_grid_lines=app_data)
    dpg.configure_item("y_axis", show_grid_lines=app_data)

# --- Ventana principal ---
with dpg.window(label="Visualizador interactivo", width=900, height=700):

    dpg.add_text("Cargar archivo CSV:")
    dpg.add_button(label="Seleccionar archivo", callback=lambda: dpg.show_item("file_dialog"))
    dpg.add_separator()

    dpg.add_button(label="Actualizar", callback=actualizar_grafico)

    dpg.add_text("Controles del grfico:")

    # Grupo horizontal para los combos
    with dpg.group(horizontal=True):
        dpg.add_text("Eje X:")
        dpg.add_combo([], tag="combo_x", width=150, callback=cambio_columna)
        dpg.add_text("Eje Y:")
        dpg.add_combo([], tag="combo_y", width=150, callback=cambio_columna)

    dpg.add_color_edit(label="Color de lnea", default_value=color_actual, callback=cambiar_color)
    dpg.add_slider_float(label="Grosor de lnea", default_value=1.0, min_value=0.5, max_value=5.0, callback=cambiar_grosor)
    dpg.add_checkbox(label="Mostrar cuadrcula", default_value=True, callback=toggle_grid)

    dpg.add_separator()

    # --- Grfico ---
    with dpg.plot(label="Datos cargados", height=450, width=-1):
        dpg.add_plot_axis(dpg.mvXAxis, label="X", tag="x_axis", no_gridlines=False)
        with dpg.plot_axis(dpg.mvYAxis, label="Y", tag="y_axis", no_gridlines=False):
            dpg.add_line_series([], [], label="Serie", tag="serie_datos", show=False)

# --- Explorador de archivos ---
with dpg.file_dialog(directory_selector=False, show=False, callback=cargar_datos,
                     id="file_dialog", width=700, height=400):
    dpg.add_file_extension(".csv", color=(150, 255, 150, 255))
    dpg.add_file_extension(".txt", color=(255, 255, 150, 255))

# --- Lanzar interfaz ---
dpg.create_viewport(title='Interfaz Interactiva - Dear PyGui', width=1000, height=800)
dpg.setup_dearpygui()
dpg.show_viewport()
dpg.start_dearpygui()
dpg.destroy_context()
```
