4. En una misma gráficar mostrar las señales S1 y S2 en función del tiempo (se sugiere usar las funciones `plt.figure()`, `plt.plot()`, `plt.title()`, `plt.legend()`, `plt.xlabel()` y `plt.ylabel()`).

````python
frecuencia_muestreo = 8000
amplitud = 1.5
frecuencia = 410
cantidad_muestras = 200
theta = np.deg2rad(40)

# COMPLETAR
````

````python
# Vector de tiempo
tiempo = np.arange(cantidad_muestras) / frecuencia_muestreo

# S1 (sinusoidal sin desfase)
s1 = amplitud * np.sin(2 * np.pi * frecuencia * tiempo)

# S2 (sinusoidal con desfase de 40 grados)
s2 = amplitud * np.sin(2 * np.pi * frecuencia * tiempo + theta)

# S1 y S2 en función del tiempo
plt.figure(figsize=(10, 6))
plt.plot(tiempo, s1, 'o-', label='Señal S1')
plt.plot(tiempo, s2, 'o-', label='Señal S2 (desfasada 40º)')
plt.title('Señales S1 y S2 en función del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# S2 vs. S1
plt.figure(figsize=(8, 8))
plt.plot(s1, s2, 'o-')
plt.title('Señal S2 vs. Señal S1')
plt.xlabel('Señal S1')
plt.ylabel('Señal S2')
plt.grid(True)
plt.show()

# Gráficas con subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: S1 y S2 en función del tiempo
ax1.plot(tiempo, s1, 'o-', label='Señal S1')
ax1.plot(tiempo, s2, 'o-', label='Señal S2 (desfasada 40º)')
ax1.set_title('Señales S1 y S2 en función del tiempo')
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Amplitud')
ax1.legend()
ax1.grid(True)

# Subplot 2: S2 vs. S1
ax2.plot(s1, s2, 'o-')
ax2.set_title('Señal S2 vs. Señal S1')
ax2.set_xlabel('Señal S1')
ax2.set_ylabel('Señal S2')
ax2.grid(True)

plt.tight_layout()
plt.show()
````

````python
# Transformada de Fourier de S1
transformada_s1 = np.fft.fft(s1)

# Vector de frecuencias
frecuencias = np.fft.fftfreq(cantidad_muestras, 1/frecuencia_muestreo)

# Frecuencias y transformada centradas en 0
frecuencias_centradas = np.fft.fftshift(frecuencias)
transformada_centrada = np.fft.fftshift(transformada_s1)

# Módulo de la transformada
modulo_transformada = np.abs(transformada_centrada)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(frecuencias_centradas, modulo_transformada, 'o-', label='Módulo de la Transformada de S1')
plt.title('Módulo de la Transformada de Fourier de S1')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud del Módulo')
plt.legend()
plt.grid(True)
plt.show()
````

## Parte 8 - Funciones
Definir una función que realice la suma ponderada mencionada en la parte anterior.   
La función recibe dos arrays y dos pesos. Dentro de la función se chequea que los dos arrays sean de la mismas dimensiones, se realiza la suma ponderada y se devuelve el array resultado. En caso que no coincide el tamaño de los arrays, la función devuelve **`None`**.

Se agregará al comienzo de la función un **`docstring`** de ayuda con la información de lo que hace la función, de los parámetros que recibe y de lo que devuelve. 




````python
def suma_ponderada(array1, array2, peso1, peso2):
    # COMPLETAR Agregar un docstring a la función.
    # El docstring se hará en formato "numpy"

    """
    Realiza la suma ponderada de dos arrays de la misma dimensión.

    Comprueba si las dimensiones de los arrays de entrada coinciden.
    Si las dimensiones coinciden, calcula la suma ponderada.
    Si no, devuelve None.

    Parameters
    ----------
    array1 : np.ndarray
        El primer array, con el que se multiplicará el peso1.
    array2 : np.ndarray
        El segundo array, con el que se multiplicará el peso2.
    peso1 : float
        El peso para el primer array.
    peso2 : float
        El peso para el segundo array.

    Returns
    -------
    np.ndarray or None
        El array resultado de la suma ponderada si las dimensiones son correctas,
        de lo contrario, devuelve None.
    """

    # COMPLETAR El código de la función

    if array1.shape == array2.shape:
        resultado = (peso1 * array1) + (peso2 * array2)
        return resultado
    else:
        return None

````
