**\# CONTESTAR**  

Para el registro **302**, 
* comentar sobre los archivos que lo componen,
* describir toda la información (metadata) del HEADER



Respuesta:<br>
    Los registros estan compuestos por tres tipos de archivos.<br>
        Header(.hea): Es un archivo de texto plano que describe el archivo de datos<br>
        Signal(.dat, .bin, .16, etc): Es el archivo de datos con las muestras de la señal sin procesar<br>
        Annotations(.atr, .ann, etc): Contiene marcadores de eventos, etiquetas o comentarios alineados con la línea de tiempo de la señal.<br><br>
    Particularmente el archivo 302.hea contiene lo siguiente:<br>
        302 2 360 510488<br>
        302.dat 212 295 12 0 -38 -12407 0 ECG<br>
        302.dat 212 213 12 0 -10 -18180 0 ECG<br><br>
    La primera linea indica: Identificador_del_registro // Cantidad_de_señales_descritas_en_el_header // Muestras_por_segundo_por_señal // Muestras_totales_por_señal<br><br>
    Las otras dos lineas serán la señales, en este cao ECG[0] y ECG[1], donde los campos que aparecen son: Nombre_del_archivo // Número_que_indica_como_se_codifico_la_señal // Factor_de_escala_para_convertir_enteros_en_unidades_fisicas_mV // Numero_de_bits_de_resolución_para_la_señal // Valor_digital_de_la_linea_de_base_cero // Valor_de_la_primera_muestra // Valor_del_checksum // Bloque_de_tiempo_de_la_señal // Nombre_de_la_señal<br>

## 2. Manejo de datos como DataFrames

Para esta parte también hay ejemplos en el notebook [demo.py de wfdb](https://github.com/MIT-LCP/wfdb-python/blob/main/demo.ipynb).


### 2.1 Pasar señales a Dataframes


```python
# COMPLETAR
# Pasar las señales a un DataFrame de pandas
record = wfdb.rdrecord('Database/physionet.org/files/stdb/1.0.0/302') # Guardo el record 302
df_signals = pd.DataFrame(record.p_signal, columns=record.sig_name) # Lo paso a dataframe de pandas
display(df_signals.head()) # Muestro las primeras 5 filas del dataframe
```

### 1.2 Graficar las señales a partir del DataFrame

```python
# COMPLETAR
# Graficar los primeros 10 segundos del DataFrame 
fs = record.fs
df_signals.iloc[0:int(10 * fs)].plot(figsize=(15, 5)) # Grafico los primeros 10s
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.title('ECG Signal from DataFrame (first 10 seconds)')
plt.show()
```

## 3. (opcional)  Detección de picos (QRS)  
 

La biblioteca **wfdb** funciones que permiten encontrar los picos (complejos QRS) de la señal de ECG.  

Para esta parte también hay ejemplos en el notebook [demo.py de wfdb](https://github.com/MIT-LCP/wfdb-python/blob/main/demo.ipynb).


### 3.1 Detecte complejos QRS sobre la primera señal del registro 302

```python
# COMPLETAR 
# Detectar los picos de la primera señal del registro 302 para los primeros 10 segundos
# Mostrar señal y picos detectados

from wfdb.processing import xqrs_detect # importo para detectar picos

record = wfdb.rdrecord('Database/physionet.org/files/stdb/1.0.0/302', sampto=int(10 * 250)) # Leo 10s del record 302

signal = record.p_signal[:, 0] # Me quedo con la primera señal
fs = record.fs

xqrs_indices = wfdb.processing.qrs.xqrs_detect(sig=signal, fs=fs) # Busco lospicos

# Graficar la señal y los picos detectados
plt.figure(figsize=(15, 5))
plt.plot(np.arange(len(signal)) / fs, signal)
plt.plot(xqrs_indices / fs, signal[xqrs_indices], 'rx', marker='o', markersize=5)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('QRS Peak Detection for Record 302 (first 10 seconds)')
plt.show()
```

### 3.2  Comparar detecciones y anotaciones

Detectar para los primeros 15 segundos y comparar con las anotaciones 

```python
# COMPLETAR
record = wfdb.rdrecord('Database/physionet.org/files/stdb/1.0.0/302', sampto=int(15 * 250)) # Cargo 15s de record 302
ann = wfdb.rdann('Database/physionet.org/files/stdb/1.0.0/302', 'atr', sampto=int(15 * 250)) # Cargo 15s de las anotaciones correspondientes
signal = record.p_signal[:, 0] # me quedo con la primera señal
fs = record.fs

xqrs_indices = wfdb.processing.qrs.xqrs_detect(sig=signal, fs=fs) # Detecto los picos

# Graficar la señal, las anotaciones y los picos detectados
plt.figure(figsize=(15, 5))
plt.plot(np.arange(len(signal)) / fs, signal, label='Señal de ECG')
plt.plot(ann.sample / fs, signal[ann.sample], 'go', label='Anotaciones', markersize=8)
plt.plot(xqrs_indices / fs, signal[xqrs_indices], 'rx', label='Picos Detectados', markersize=8)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Comparison of Detected Peaks and Annotations (first 15 seconds)')
plt.legend()
plt.show()


```
